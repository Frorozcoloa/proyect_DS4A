```python
from pygram import PyGram
import pandas as pd
from datetime import datetime


pygram = PyGram()
```

```python

username = 'offcorss.us'

#Se crea una función que extraiga el DF ó posts_list
def get_posts_list(username, limit=100):
    
    """Return of list of dict"""
    posts = pygram.get_posts(username, limit=limit)
    
    posts_list = []
    for i,post in enumerate(posts):
        posts_list.append(post)
        print(f'Scrapy post {i+1}')

    # posts_df = pd.DataFrame(posts_list)
    # posts_df['Date Create'] = posts_df.timestamp.apply(lambda x: datetime.fromtimestamp(x))
    # posts_df.drop('timestamp', axis=1, inplace=True)
    return posts_list
```

```python
list_posts =  get_posts_list(username, limit=100)
```

```python
def get_comments_list(posts_list, limit=100):
    """Return list of dict"""
    comments_list = []
    
    
    for i,post in enumerate(posts_list):
        
        comments = pygram.get_comments(post, limit=limit)
        
        print(f'Start Scrapy post {i+1}')
        
        
        for i ,comment in enumerate(comments):
            if comment:
                comments_list.append(comment)
                print(f'\tScrapy comment {i+1}')
        
    
    return comments_list
            
        
        
    
```
