```python
import facebook_scraper
from facebook_scraper import get_posts, page_iterators
import pandas as pd
```

```python
pages = [ 'PolitoColombia']
list_post_reaction = []

for page in pages:
    print(f'Iniciando page {page.upper()}')
    
    try: #Existencia de la página o la facebook te saca # id:1
    
        get_post = get_posts(page, pages=15000, extra_info=True)
        for i,post in enumerate(get_post):
    
    
            try: # Mira si el post tiene reacciones  # id:2
                a = post['reactions'] 
                print('Hay Reacción')
                post['Have_reaction'] = True
            except: # id:2
                post['reactions'] = None # en caso de que hay post, lo crea y ponr None
                post['Have_reaction'] = False
                print('No hay reaccion')
    
    
            list_post_reaction.append(post)
    
    
            if i% 50 == 0: #condoción para que guarde los datos cada 50 iter
                print()
                print(f'Iniciando guardado')
        
                df_post_reaction = pd.DataFrame(list_post_reaction)
            
            
                try: #miramos si existe la base datos creada #id:3
                    #read
                    df_post_db = pd.read_csv('database_post_facebook.csv')
                    #procesamiento de concatenación
                    df_post_db = pd.concat([df_post_reaction, df_post_db])
                    df_post_db['post_id'] = df_post_db['post_id'].astype('int64')
                    df_post_db.drop_duplicates(subset = 'post_id', keep = 'last', inplace = True)
                    #save
                    df_post_db.to_csv('database_post_facebook.csv', index= False)
                except: #id:3
                    #Se crea la base de datos
                     df_post_reaction.to_csv('database_post_facebook.csv', index= False)
        
    except: #id:1
        print(f'Ha habido un error en la page {page}')
        print('Se continurá con la siguiente')
        
```
